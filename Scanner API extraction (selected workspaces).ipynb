```python
# 2024-07-23 B.Agullo - Initial version

# This script will get the workspace info (all items, with properties) for a given workspace id
# to access the power query expressions and schema you need to have enhanced API responses enabled in the tenant

import time
import datetime
import sempy.fabric as fabric 
import json
import os

today = datetime.datetime.now()
apiName = "WorkspaceInfo"

dt = today.strftime("%y%m%d")
y = today.strftime("%y")
m = today.strftime("%m")
d = today.strftime("%d")
file_name = f"workspaceInfo_{dt}.json"

folder_path = f"/lakehouse/default/Files/API/{apiName}/{y}/{m}/{d}/"


if not os.path.exists(folder_path):
    os.makedirs(folder_path)

file_path = os.path.join(folder_path, file_name)


client = fabric.FabricRestClient()

fabric_api_url = "v1.0/myorg/admin/workspaces/getInfo?lineage=True&datasourceDetails=True&datasetSchema=True&datasetExpressions=True"


#update as needed
body = {
  "workspaces": [
    "57f939a1-89e4-4447-851c-b8decb063a5d"
  ]
}

response = client.post(fabric_api_url, json = body)

json_data = response.json()

scanId = json_data["id"]

check = True

while check:
    response = client.get(f"v1.0/myorg/admin/workspaces/scanStatus/{scanId}" ) 
    json_data = response.json()
    status = json_data["status"]

    print(status)

    check = (status != "Succeeded")
    time.sleep(5)


response = client.get(f"v1.0/myorg/admin/workspaces/scanResult/{scanId}")
json_data = response.json()




with open(file_path, 'w') as json_file:
    json.dump(json_data, json_file, indent= 4)

print("File saved")
```
